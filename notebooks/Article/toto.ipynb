```python
import bs4
from urllib.request import Request, urlopen
import re
import pandas as pd
```

```python
req = Request('https://www.bigdataparis.com/les-speakers?page=1#guests-list-70035')
html_page = urlopen(req).read()
```

```python
d = bs4.BeautifulSoup(html_page, 'html.parser')
```

```python
all = d.find_all("div", class_="caption")
```

```python
all = 
for i in all:
    print(i.text)
```

```python
res = pd.DataFrame(columns=["Name", "Title", "Company", "linkedin"])
for i in d.find_all("div", class_="col-md-4 col-xs-6 guest-item"):
    h = []
    for j in i.find("div", class_="caption"):    
        txt = j.text
        txt = txt.strip().split("\n")
        for e in range(len(txt)):
            txt[e] = txt[e].strip()
        txt = list(filter(None, txt))
        if txt != []:
            h.append(txt)

    print(h)
    #if len(txt) == 3:
    #    res = res.append({"Name": txt[0], "Title": txt[1], "Company": txt[2]}, ignore_index=True)
    #else:
    #    res = res.append({"Name": txt[0], "Title": txt[1], "Company": ""}, ignore_index=True)
    
    if len(h) == 2:
        h = h.append([])
    
    for a in i.find_all("a"):
        pp = Request(f'https://www.bigdataparis.com/{a["href"].strip()}')
        dd = bs4.BeautifulSoup(urlopen(pp).read(), 'html.parser')
        for kk in dd.find_all("a", class_="linkedin"):
            h.append(kk["href"])
            
    res.append(h)
    
print(res)
    
```
