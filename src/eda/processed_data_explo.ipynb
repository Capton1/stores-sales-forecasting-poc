## Trusted

```python
train_set = pd.read_csv(f"data/trusted/training_set.csv", index_col=0)
val_set = pd.read_csv(f"data/trusted/validation_set.csv", index_col=0)

# train_set = train_set[['sales', 'onpromotion',
#       'dcoilwtico', 'cluster', 'Holiday', 'Normal', 'Additional', 'Event']]
train_set.drop("state", inplace=True, axis=1)
val_set.drop("state", inplace=True, axis=1)

train_set
```

```python
def get_features_and_target(df: pd.DataFrame, target="sales"):
    """get features and target sets for model training and eval

    Args:
        df (pd.DataFrame): df with features and y
        target (str): target column name

    Returns:
        Tuple[pd.DataFrame, pd.DataFrame]: features_set, target_set
    """
    return df.drop(["date", target], axis=1), df[[target]]
```

```python
X_train, y_train = get_features_and_target(train_set)
X_val, y_val = get_features_and_target(val_set)

X_train.shape, y_train.shape
```

```python
X_train = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))
y_train = y_train.values.reshape((y_train.shape[0], 1, y_train.shape[1]))
```

```python
X_val = X_val.values.reshape((X_val.shape[0], 1, X_val.shape[1]))
y_val = y_val.values.reshape((y_val.shape[0], 1, y_val.shape[1]))
```

```python
from keras.models import Sequential, load_model
from keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(100, activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam", loss="mse")
```

```python
history = model.fit(
    X_train, y_train, epochs=30, batch_size=50, validation_data=(X_val, y_val)
)
```
